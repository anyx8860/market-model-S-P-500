## api get-movers

```python


url = "https://apidojo-yahoo-finance-v1.p.rapidapi.com/market/get-movers"

querystring = {"region":"US","lang":"en-US"}#,"start":"0"#,"count":"6"}

headers = {
    'x-rapidapi-key': "###########################################",
    'x-rapidapi-host': "apidojo-yahoo-finance-v1.p.rapidapi.com"
    }

response = requests.request("GET", url, headers=headers, params=querystring).json()

print(json.dumps(response, indent=4, sort_keys = True))
```

```python
response['finance']#['result'][0]#['quotes'][0]#['rawCriteria']#['offset']#['criteriaMeta']#['criteria']
```

```python
id_ = []
title = []
description = []
canonicalName = []
size = []
offset = []
sortField = []
sortType = []
quoteType = []
topOperator = []
field = []
operators = []
values = []
labelsSelected = []
rawCriteria = []
start = []
count = []
total = []
language = []
region = []
quoteType_quotes = []
quoteSourceName = []
triggerable = []
priceHint = []
firstTradeDateMilliseconds = []
exchange = []
market = []
fullExchangeName = []
marketState = []
sourceInterval = []
exchangeDataDelayedBy = []
exchangeTimezoneName = []
exchangeTimezoneShortName = []
gmtOffSetMilliseconds = []
esgPopulated = []
tradeable = []
symbol = []
predefinedScr = []
versionId = []
error = []
for res in response['finance']['result']:
    for crit in res['criteriaMeta']['criteria']:
        for opts in crit['operators']:
            for val in crit['values']:
                if(0 < len(crit['labelsSelected'])):
                    l = len(crit['labelsSelected'])
                else:
                    l = 1
                for ls in range(l):
                    for q in res['quotes']:
                        
                        id_.append(res['id'])
                        title.append(res['title'])
                        description.append(res['description'])
                        canonicalName.append(res['canonicalName'])
                        size.append(res['criteriaMeta']['size'])
                        offset.append(res['criteriaMeta']['offset'])
                        sortField.append(res['criteriaMeta']['sortField'])
                        sortType.append(res['criteriaMeta']['sortType'])
                        quoteType.append(res['criteriaMeta']['quoteType'])
                        topOperator.append(res['criteriaMeta']['topOperator'])
                        field.append(crit['field'])
                        operators.append(opts)
                        values.append(val)
                        try:
                            labelsSelected.append(crit['labelsSelected'][ls])
                        except:
                            labelsSelected.append(np.nan)
                        rawCriteria.append(res['rawCriteria'])
                        start.append(res['start'])
                        count.append(res['count'])
                        total.append(res['total'])
                        language.append(q['language'])
                        region.append(q['region'])
                        try:
                            quoteType_quotes.append(q['quoteType'])
                        except:
                            quoteType_quotes.append(np.nan)
                        quoteSourceName.append(q['quoteSourceName'])
                        triggerable.append(q['triggerable'])
                        priceHint.append(q['priceHint'])
                        firstTradeDateMilliseconds.append(q['firstTradeDateMilliseconds'])
                        exchange.append(q['exchange'])
                        market.append(q['market'])
                        fullExchangeName.append(q['fullExchangeName'])
                        marketState.append(q['marketState'])
                        sourceInterval.append(q['sourceInterval'])
                        exchangeDataDelayedBy.append(q['exchangeDataDelayedBy'])
                        exchangeTimezoneName.append(q['exchangeTimezoneName'])
                        exchangeTimezoneShortName.append(q['exchangeTimezoneShortName'])
                        gmtOffSetMilliseconds.append(q['gmtOffSetMilliseconds'])
                        esgPopulated.append(q['esgPopulated'])
                        tradeable.append(q['tradeable'])
                        symbol.append(q['symbol'])
                        predefinedScr.append(res['predefinedScr'])
                        versionId.append(res['versionId'])
                        error.append(response['finance']['error'])
    


df = pd.DataFrame({'id': id_, 'title': title, 'description': description, 'canonicalName': canonicalName, 
            'size': size, 'offset': offset, 'sortField': sortField, 'sortType': sortType, 'quoteType': quoteType,
    'topOperator': topOperator, 'field': field, 'operators': operators, 'values': values, 
    'labelsSelected': labelsSelected, 'rawCriteria': rawCriteria, 'start': start, 'count': count, 'total': total,
    'language': language, 'region': region, 'quoteType_quotes': quoteType_quotes, 'quoteSourceName': quoteSourceName,
    'triggerable': triggerable, 'priceHint': priceHint, 'firstTradeDateMilliseconds': firstTradeDateMilliseconds,
                  'exchange': exchange, 'fullExchangeName': fullExchangeName, 'marketState': marketState, 
                  'sourceInterval': sourceInterval, 'exchangeDataDelayedBy': exchangeDataDelayedBy,
                  'exchangeTimezoneName': exchangeTimezoneName, 
    'exchangeTimezoneShortName': exchangeTimezoneShortName, 'gmtOffSetMilliseconds': gmtOffSetMilliseconds,
                  'esgPopulated': esgPopulated, 'tradeable': tradeable, 'symbol': symbol, 
                   'predefinedScr': predefinedScr, 'versionId': versionId, 'error': error})
df.head()
```

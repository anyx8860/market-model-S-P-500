# Here we will be webscraping the inflation data from [US Inflation Calculator](https://www.usinflationcalculator.com/inflation/current-inflation-rates/)

getting webpage

```python
import requests
from bs4 import BeautifulSoup as bs
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
# PyMySQL 
import pymysql
pymysql.install_as_MySQLdb()
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
from sqlalchemy import desc
import mysql.connector as sql
import pandas as pd
from datetime import datetime as dt
import numpy as np

r = requests.get("https://www.usinflationcalculator.com/inflation/current-inflation-rates/")
webpage = bs(r.content)
print(webpage.prettify())
```

### getting table from website

```python
table = webpage.select('table')
table
```

```python
table[0].select('tr')[1]#.select('td')[0].select('strong')[0].text
```

```python
table[0].select('tr')[1].select('td')
```

### getting data into list

```python
year = []
month = []
inflation = []
for tr in table[0].select('tr')[1:(len(table[0].select('tr')) - 1)]:
    y = tr.select('td')[0].select('strong')[0].text
    r  = range(1, len(tr.select('td')) - 1)
    for i in r:
        year.append(y)
        month.append(str(i))
        try:
            inflation.append(float(tr.select('td')[i].text))
        except:
            inflation.append(np.nan)
    
```

**checking data**

```python
assert (len(inflation) == len(year)) and (len(year) == len(month))
```

### making dataframe

```python
df  = pd.DataFrame({'year': year, 'month': month, 'inflation_%': inflation})
df.head()
```

### Loading to database

```python
mydb = create_engine("mysql://Owner:######@localhost/market_database")
df.to_sql(name = 'inflation_table', con = mydb, schema= 'market_database', if_exists = 'replace')
```

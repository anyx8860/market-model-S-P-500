### econmic indecator table

```python
import requests
from bs4 import BeautifulSoup as bs
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
# PyMySQL 
import pymysql
pymysql.install_as_MySQLdb()
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
from sqlalchemy import desc
import mysql.connector as sql
import pandas as pd
from datetime import datetime as dt
import numpy as np

mydb = create_engine("mysql://Owner:########@localhost/market_database")
df_indecators = pd.read_sql("select * from maket_indicators_table", con = mydb).drop(['index'], axis=1)
df_indecators.head()
```

### s&p 500 data

```python
df_market = pd.read_sql("select * from market_index_table_quarter", con = mydb)#.drop(['index'], axis=1)
df_market.head()
```

### merging data

```python
df = pd.merge(df_market, df_indecators, on=['date'], how='inner'
             ).sort_values(by=['index']).drop('index', axis=1).drop_duplicates()

df.head()
```

### loading data

```python
mydb = create_engine("mysql://Owner:########@localhost/market_database")
df.to_sql(name = 'market_table_3', con = mydb, schema= 'market_database', if_exists = 'replace')
```
